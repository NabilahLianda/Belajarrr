```python
import pandas as pd
from tabulate import tabulate

# URL Google Sheet yang dipublikasikan sebagai CSV
csv_url = "https://docs.google.com/spreadsheets/d/1K_BU1IPKznoAmAtY4nL3KglrZ7FZQ7JWgeTtyublWzs/export?format=csv"

# Membaca data dari URL CSV ke DataFrame
data = pd.read_csv(csv_url)

# Menghapus baris dan kolom yang kosong
data_cleaned = data.dropna(how='all')  # Hapus baris kosong
data_cleaned = data_cleaned.dropna(axis=1, how='all')  # Hapus kolom kosong

# Menampilkan tabel dengan format grid
print(tabulate(data_cleaned, headers="keys", tablefmt="grid"))
```

```python
#User Memasukkan Data

print(f"Masukkan Data Diri Anda:\n")
pekerjaan = input("Pekerjaan (Wiraswasta/Tidak Bekerja): ")
usia = input("Usia (20-29/30-40): ")
status = input("Status (Kawin/Belum Kawin): ")
penghasilan = input("Penghasilan (Kurang Dari 1 Juta/2 - 3 Juta/4 - 5 Juta/Lebih Dari 5 Juta): ")
kendaraan = input("Kendaraan (Motor/Mobil/Angkutan Umum): ")
kepemilikan = input("Kepemilikan rumah (Menyewa/Orang Tua/Pribadi): ")
atap = input("Jenis atap bangunan (Asbes/Genteng): ")

#Data Pekerjaan User
if pekerjaan == 'Wiraswasta':
  PekerjaanL = prob_wiraswastaL
  PekerjaanTL = prob_wiraswastaTL
elif pekerjaan == 'Tidak Bekerja':
  PekerjaanL = prob_TbekerjaL
  PekerjaanTL = prob_TbekerjaTL

#Data Usia User
if usia == '20-29':
  UsiaL = prob_mudaL
  UsiaTL = prob_mudaTL
elif usia == '30-40':
  UsiaL = prob_tuaL
  UsiaTL = prob_tuaTL

#Data Status User
if status == 'Kawin':
  StatusL = prob_kawinL
  StatusTL = prob_kawinTL
elif status == 'Belum Kawin':
  StatusL = prob_BkawinL
  StatusTL = prob_BkawinTL

#Data Penghasilan User
if penghasilan == 'Kurang Dari 1 Juta':
  GajiL = prob_G1L
  GajiTL = prob_G1TL
elif penghasilan == '2 - 3 Juta':
  GajiL = prob_G2L
  GajiTL = prob_G2TL
elif penghasilan == '4 - 5 Juta':
  GajiL = prob_G3L
  GajiTL = prob_G3TL
elif penghasilan == 'Lebih Dari 5 Juta':
  GajiL = prob_G4L
  GajiTL = prob_G4TL

#Data Kendaraan User
if kendaraan == 'Mobil':
  KendaraanL = prob_mobilL
  KendaraanTL = prob_mobilTL
elif kendaraan == 'Motor':
  KendaraanL = prob_motorL
  KendaraanTL = prob_motorTL
elif kendaraan == 'Angkutan Umum':
  KendaraanL = prob_AumumL
  KendaraanTL = prob_AumumTL

#Data Kepemilikan User
if kepemilikan == 'Pribadi':
  KepemilikanL = prob_pribadiL
  KepemilikanTL = prob_pribadiTL
elif kepemilikan == 'Orang Tua':
  KepemilikanL = prob_ortuL
  KepemilikanTL = prob_ortuTL
elif kepemilikan == 'Menyewa':
  KepemilikanL = prob_sewaL
  KepemilikanTL = prob_sewaTL

#Data Atap Rumas User
if atap == 'Asbes':
  AtapL = prob_asbesL
  AtapTL = prob_asbesTL
elif atap == 'Genteng':
  AtapL = prob_gentengL
  AtapTL = prob_gentengTL

#Menghitung Seluruh Kelas
prob_layak = PekerjaanL * UsiaL * StatusL * GajiL * KendaraanL * KepemilikanL * AtapL
prob_Tlayak = PekerjaanTL * UsiaTL * StatusTL * GajiTL * KendaraanTL * KepemilikanTL * AtapTL

print(f"Hasil Probabilitas:\n")
print(f"Probabilitas Layak: {prob_layak}")
print(f"Probabilitas Tidak Layak: {prob_Tlayak}")

if prob_layak > prob_Tlayak:
   prediksi = "Layak"
elif prob_Tlayak > prob_layak:
   prediksi = "Tidak Layak"

print(f"\nHasil Kelayakan: {prediksi}")
```
